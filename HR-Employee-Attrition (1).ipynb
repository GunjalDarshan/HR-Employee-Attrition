```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the aesthetics for the plots
sns.set(style="whitegrid")

# Create a figure with subplots
fig, axes = plt.subplots(2, 2, figsize=(18, 12))

# Age distribution with respect to Attrition
sns.histplot(data=df, x="Age", hue="Attrition", multiple="stack", ax=axes[0, 0], kde=True)
axes[0, 0].set_title("Age Distribution by Attrition")

# Monthly Income distribution with respect to Attrition
sns.histplot(data=df, x="MonthlyIncome", hue="Attrition", multiple="stack", ax=axes[0, 1], kde=True)
axes[0, 1].set_title("Monthly Income Distribution by Attrition")

# Years at Company distribution with respect to Attrition
sns.histplot(data=df, x="YearsAtCompany", hue="Attrition", multiple="stack", ax=axes[1, 0], kde=True)
axes[1, 0].set_title("Years at Company by Attrition")

# OverTime effect on Attrition
sns.countplot(x="OverTime", hue="Attrition", data=df, ax=axes[1, 1])
axes[1, 1].set_title("Effect of OverTime on Attrition")

plt.tight_layout()
plt.show()
```

```python
# Create a DataFrame for counts of Attrition by EducationField
attrition_by_education = df.groupby(['EducationField', 'Attrition']).size().unstack(fill_value=0)

# Create a donut plot for Attrition by EducationField
fig, ax = plt.subplots(figsize=(8, 8))
attrition_by_education.plot(kind='pie', y='Yes', legend=False, ax=ax, autopct='%1.1f%%', startangle=140, pctdistance=0.85, labels=attrition_by_education.index)

# Draw a circle at the center of pie to make it look like a donut
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

# Equal aspect ratio ensures that pie is drawn as a circle
ax.axis('equal')  
plt.tight_layout()
plt.title('Attrition by Education Field')
plt.show()
```

```python
# Create a DataFrame for counts of Attrition by Age
attrition_by_age = df.groupby(['Age', 'Attrition']).size().unstack(fill_value=0)

# Plot Attrition by Age
fig, ax = plt.subplots(figsize=(12, 6))
attrition_by_age.plot(kind='bar', stacked=True, ax=ax, color=['#5fba7d', '#d9534f'])
ax.set_title('Attrition by Age')
ax.set_xlabel('Age')
ax.set_ylabel('Number of Employees')
ax.legend(title='Attrition', labels=['No', 'Yes'])

plt.tight_layout()
plt.show()
```

```python
# Create a DataFrame for counts of Attrition by JobRole
attrition_by_jobrole = df.groupby(['JobRole', 'Attrition']).size().unstack(fill_value=0)

# Plot Attrition by JobRole
fig, ax = plt.subplots(figsize=(12, 8))
attrition_by_jobrole.plot(kind='barh', stacked=True, ax=ax, color=['#5fba7d', '#d9534f'])
ax.set_title('Attrition by Job Role')
ax.set_xlabel('Number of Employees')
ax.set_ylabel('Job Role')
ax.legend(title='Attrition', labels=['No', 'Yes'])

plt.tight_layout()
plt.show()
```

```python
# Assuming "YearsAtCompany" as an approximation for the year dimension in this context,
# since actual join and leave years are not directly available in the dataset.

# Group by YearsAtCompany and calculate attrition rate
attrition_rate_by_years = df.groupby('YearsAtCompany')['Attrition'].apply(lambda x: (x == 'Yes').mean()).reset_index()

# Plot the trend of attrition rate by Years at Company
fig, ax = plt.subplots(figsize=(12, 6))
sns.lineplot(data=attrition_rate_by_years, x='YearsAtCompany', y='Attrition', ax=ax, marker='o')
ax.set_title('Trend of Attrition Rate by Years at Company')
ax.set_xlabel('Years at Company')
ax.set_ylabel('Attrition Rate')

plt.xticks(range(0, attrition_rate_by_years['YearsAtCompany'].max() + 1, 1))
plt.tight_layout()
plt.show()
```

```python
# Add total counts to the attrition_counts_by_jobrole for visualization
attrition_by_jobrole['Total'] = attrition_by_jobrole.sum(axis=1)

# Sort by Total to see which Job Roles have the most attrition
attrition_counts_by_jobrole_sorted = attrition_by_jobrole.sort_values(by='Total', ascending=False)

# Plot Attrition by JobRole with Total Counts
fig, ax = plt.subplots(figsize=(14, 9))
attrition_counts_by_jobrole_sorted[['No', 'Yes']].plot(kind='barh', stacked=True, ax=ax, color=['#5fba7d', '#d9534f'])
ax.set_title('Attrition Counts by Job Role with Total Counts')
ax.set_xlabel('Number of Employees')
ax.set_ylabel('Job Role')
ax.legend(title='Attrition', labels=['No', 'Yes'])

# Add total counts as text labels on the bars
for index, value in enumerate(attrition_counts_by_jobrole_sorted['Total']):
    ax.text(value, index, ' ' + str(value), va='center', ha='left', fontsize=10)

plt.tight_layout()
plt.show()
```

```python
# Get unique job roles
unique_job_roles = df['JobRole'].unique()

# Set up the matplotlib figure layout
n_cols = 2
n_rows = (len(unique_job_roles) + n_cols - 1) // n_cols
fig, axes = plt.subplots(n_rows, n_cols, figsize=(16, 4 * n_rows), sharey=True)

# Flatten axes array for easy iteration
axes = axes.flatten()

for index, job_role in enumerate(unique_job_roles):
    # Filter the dataset for the current job role
    job_role_df = df[df['JobRole'] == job_role]
    
    # Calculate the attrition counts for the current job role
    job_role_attrition_counts = job_role_df['Attrition'].value_counts()
    
    # Add total count
    job_role_attrition_counts['Total'] = job_role_attrition_counts.sum()

    # Plot the counts for the current job role
    job_role_attrition_counts.drop('Total').plot(kind='bar', ax=axes[index], color=['#2ca02c', '#d62728'])
    axes[index].set_title(f'Attrition Counts for {job_role}')
    axes[index].set_xlabel('Attrition')
    axes[index].set_ylabel('Count')
    axes[index].grid(axis='y')
    
    # Add total count as text on the plot
    axes[index].text(1, job_role_attrition_counts['Total']/2, 'Total: ' + str(job_role_attrition_counts['Total']), ha='center', va='bottom', fontsize=12)

# Hide any unused axes if the number of job roles is odd
for index in range(len(unique_job_roles), len(axes)):
    fig.delaxes(axes[index])

plt.tight_layout()
plt.show()
```
